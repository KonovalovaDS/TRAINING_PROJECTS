## <center> Forecasting gas production at wells </center>

Mr. Smith is the founder of the company named "Oil & Gas", which has already run 200 wells of gas production. This year Mr. Smith is opening 30 new wells. However, in order to optimize costs and increase income, he needs to estimate how much money each of the wells will bring him, as well as understand which factors (well parameters) have the greatest potential impact on the volume of gas production. To do this, Mr. Smith decided to Data Science techniques. 

Mr. Smith has a set of historical data showing the gas production at his wells.

**Features in the data:**

- Well — well identifier;

- Por — well porosity (%);

- Perm — well permeability;

- AI — acoustic impedance ($kg/m^2 * 10^6$);

- Brittle — well brittleness coefficient (%);

- TOC — total organic carbon (%);

- VR — vitrinite reflectivity (%);

- Prod — daily gas production (million cubic feet).

**Goal of the Project**: to build a regression model that predicts gas production at a well (target feature — *Prod*) based on other well characteristics, and interpret the results of your model.

The Project is split into two parts:

- In the first, to build a simple linear regression model, analyze its results, and select the most significant factors for forecasting.

- In the second, to build a polynomial regression model with regularization and look at the final modeling results.

### Linear Regression Model build using Numpy library and OLS

````python
factors = data[['Well', 'Por', 'Perm', 'AI', 'Brittle', 'TOC', 'VR']]
target = data[['Prod']]

X = np.column_stack((np.ones(200), factors))
y = target

w_hat = np.linalg.inv(X.T @ X) @ X.T @ y
print(np.round(w_hat.values, 0))
````

#### Conclusion

> The "well identifier" factor is not informative.

> For example, if the Well Fragility Factor changes by one percent, the Daily Gas Production will increase by 25 million cubic feet. At the same time, changing the Total Organic Carbon by one percent will reduce the Daily Gas Production by 78 million cubic feet.

#### Update the model

````python
features = data[['Por', 'AI', 'Brittle', 'VR']]
target = data[['Prod']]

A = np.column_stack((np.ones(200), features))
y = target

print(A.shape)
print('\n')

w_hat = np.linalg.inv(A.T @ A) @ A.T @ y
print(np.round(w_hat.values, 0))
print('\n')

y_pred = np.dot(A, w_hat)

print('MAE {:.2f}:'.format(metrics.mean_absolute_error(y, y_pred)))
print('MAPE {:.2f}%:'.format(metrics.mean_absolute_percentage_error(y, y_pred) * 100))
````

#### Conclusion

> All features whose coefficients contradicted the correlation were excluded. However, the model error increased by 0.41 percentage points, which in absolute terms resulted in an increase in the model error by almost 18 MMcf/d of gas production compared to the target indicators. As for individual coefficients, they also changed. For example, an increase in the well fragility coefficient by one percent will lead to an increase in gas production by 28 MMcf/d, but in the previous calculation this increase was 25 MMcf/d. An increase in the absolute equivalent of Acoustic Impedance will lead to a decrease in gas production by 200 MMcf/d, but in the previous model run this figure was 365.

### Linear Regression Model build using Scikit-Learn Library

````python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate

X = data.drop(['Well', 'Perm', 'TOC', 'Prod'], axis = 1)
y = data['Prod']

model = LinearRegression(fit_intercept = False)
model.fit(X, y)

print('w_hat:', np.round(model.coef_, 0))
print('\n')

y_pred = model.predict(X)

print('MAE {:.2f}:'.format(metrics.mean_absolute_error(y, y_pred)))
print('MAPE {:.2f}%:'.format(metrics.mean_absolute_percentage_error(y, y_pred) * 100))
print('\n')

cv_result = cross_validate(model, X, y, scoring = 'neg_mean_absolute_percentage_error', cv = 5, return_train_score = True)

print('MAPE on train folds: {:.2f}%'.format(-cv_result['train_score'].mean()*100))
print('MAPE on validation folds: {:.2f}%'. format(-cv_result['test_score'].mean()*100))
````

#### Conclusion

> The model built with the sklearn library showed a result one percentage point worse than the model built manually. In all likelihood, this happened because the library has a built-in singular decomposition function. In the future, we need to try to standardize the data and use polynomial regression and regularization.

### Polynomial Regression Model and Regularization

The target is to reduce the value of the model error. 

````python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
````

````python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

poly = PolynomialFeatures(degree = 3, include_bias = False)
X_poly = poly.fit_transform(X_scaled)
print(X_poly.shape)

lr = LinearRegression()

cv_result = cross_validate(lr, X_poly, y, scoring = 'neg_mean_absolute_percentage_error', cv = 5, return_train_score = True)

print('\n')
print('MAPE on train folds: {:.2f}%'.format(-cv_result['train_score'].mean()*100))
print('MAPE on validation folds: {:.2f}%'. format(-cv_result['test_score'].mean()*100))
````

#### Conclusion

> The linear regression model on polynomial features improved the metric on the validation sample almost twice.

#### Linear Regression with $L_1$-regularization

````python
param_distribution = {'alpha' : list(np.linspace(0.1, 1, 10, dtype = float))}

random_search = RandomizedSearchCV(estimator = Lasso(max_iter = 10000),
                                  param_distributions = param_distribution,
                                  cv = 5, 
                                  n_iter = 10,
                                  n_jobs = -1)
random_search.fit(X_poly, y)

print('The best value of hyperparameters: {}'.format(random_search.best_params_))
````

````python
lasso = Lasso(alpha = 1, max_iter = 10000)

cv_result = cross_validate(lasso, X_poly, y, scoring = 'neg_mean_absolute_percentage_error', cv = 5, return_train_score = True)

print('MAPE on train folds: {:.2f}%'.format(-cv_result['train_score'].mean()*100))
print('MAPE on validation folds: {:.2f}%'. format(-cv_result['test_score'].mean()*100))
````

#### Conclusion

> The metric improved by 0.2 percentage points.

#### Linear Regression with $L_2$-regularization

````python
param_distribution = {'solver':['auto', 'svd', 'sag', 'saga'],
                      'alpha' : list(np.linspace(0.1, 1, 1, dtype = float))}

grid_search = GridSearchCV(estimator = Ridge(max_iter = 10000),
                                  param_grid = param_distribution,
                                  cv = 5, 
                                  n_jobs = -1)
grid_search.fit(X_poly, y)

print('The best value of hyperparameters: {}'.format(grid_search.best_params_))
````

````python
ridge = Ridge(alpha = 0.1, solver = 'svd')

cv_results = cross_validate(ridge, X_poly, y, scoring = 'neg_mean_absolute_percentage_error', cv = 5, return_train_score = True)

print('MAPE on train folds: {:.2f}%'.format(-cv_results['train_score'].mean()*100))
print('MAPE on validation folds: {:.2f}%'.format(-cv_results['test_score'].mean()*100))
````

#### Conclusion

> The validation data metric has deteriorated, while the training data metric has remained unchanged.

#### Linear Regression with combination of $L_1$- & $L_2$-regularization

````python
param_distribution = {'l1_ratio':list(np.linspace(0.1, 1, 1, dtype = float)),
                      'alpha' : list(np.linspace(0.1, 1, 1, dtype = float))}

grid_search = GridSearchCV(estimator = ElasticNet(max_iter = 10000),
                                  param_grid = param_distribution,
                                  cv = 5, 
                                  n_jobs = -1)
grid_search.fit(X_poly, y)

print('The best value of hyperparameters: {}'.format(grid_search.best_params_))
````

````python
elasticnet = ElasticNet(alpha = 0.1, l1_ratio = 0.1)

cv_results = cross_validate(elasticnet, X_poly, y, scoring = 'neg_mean_absolute_percentage_error', cv = 5, return_train_score = True)

print('MAPE on train folds: {:.2f}%'.format(-cv_results['train_score'].mean()*100))
print('MAPE on validation folds: {:.2f}%'.format(-cv_results['test_score'].mean()*100))
````

#### Conclusion

> All metrics have deteriorated significantly.
